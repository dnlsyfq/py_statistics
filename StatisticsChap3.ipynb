```python
import pandas as pd 
import matplotlib.pyplot as plt
from scipy import stats
import numpy as np
```

```python
df = pd.read_csv('wnba.csv')
```

```python
df['Experience'].value_counts().plot.bar()
```

```python
df['Experience'] = df['Experience'].apply(lambda x: 0 if x == 'R'else x)
```

```python
df['Experience'].value_counts().plot.bar()
```

```python
df['Experience'] = df['Experience'].astype('int')
```

```python
def exp_label(row):
    if row['Experience'] >= 10:
        return 'Veteran'
    elif row['Experience'] >= 5 and row['Experience'] < 10:
        return 'Very experienced'
    elif row['Experience'] >= 4 and row['Experience'] < 5:
        return 'Experienced'
    elif row['Experience'] >= 1 and row['Experience'] < 4:
        return 'Little experience'
    else:
        return 'Rookie'
```

```python
df['Exp_ordinal'] = df.apply(exp_label,axis=1)
```

```python
df['Exp_ordinal'].value_counts().\
loc[['Rookie','Little experience','Experienced','Very experienced','Veteran']].\
plot.barh(title='Experiences Players')
```

```python
df['Exp_ordinal'].value_counts().\
loc[['Rookie','Little experience','Experienced','Very experienced','Veteran']].\
plot.bar(title='Experiences Players',rot=45)
```

```python
df['Pos'].value_counts().plot.pie()
```

```python
exp_percent = df['Exp_ordinal'].value_counts(normalize=True)*100
exp_percent.plot.pie(figsize=(6,6),autopct='%.1f%%')
```

'%.1f%%'

* %  announces string formatting
* '.1' mentions the decimal precision; if we wanted a precision of 5 decimal places, we'd need to write '.5'
* f displays a no. as a fixed-point number
* % mentions that we want to use percentages
* % includes the percentage symbol in string that will be displayed
    on the pie chart

```python
df['Exp_ordinal'].value_counts().\
plot.pie(\
         figsize=(6,6),
         autopct='%.2f%%',
         title="Percentage of players in WNBA by level of experience")
plt.ylabel('')
```
