```python
import pandas as pd 
```

```python
df = pd.read_csv('wnba.csv')
```

```python
df['Experience'].value_counts().plot.bar()
```

```python
df['Experience'] = df['Experience'].apply(lambda x: 0 if x == 'R'else x)
```

```python
df['Experience'].value_counts().plot.bar()
```

```python
df['Experience'] = df['Experience'].astype('int')
```

```python
def exp_label(row):
    if row['Experience'] >= 10:
        return 'Veteran'
    elif row['Experience'] >= 5 and row['Experience'] < 10:
        return 'Very experienced'
    elif row['Experience'] >= 4 and row['Experience'] < 5:
        return 'Experienced'
    elif row['Experience'] >= 1 and row['Experience'] < 4:
        return 'Little experience'
    else:
        return 'Rookie'
```

```python
df['Exp_ordinal'] = df.apply(exp_label,axis=1)
```

```python
df['Exp_ordinal'].value_counts().\
loc[['Rookie','Little experience','Experienced','Very experienced','Veteran']].\
plot.barh(title='Experiences Players')
```

```python
df['Exp_ordinal'].value_counts().\
loc[['Rookie','Little experience','Experienced','Very experienced','Veteran']].\
plot.bar(title='Experiences Players',rot=45)
```
